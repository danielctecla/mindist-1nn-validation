```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
whiteWine = pd.read_csv('whine-data/winequality-white.csv', sep=';')
redWine = pd.read_csv('whine-data/winequality-red.csv', sep=';')
```

```python
# Declare class
whiteWine['class'] = 0  # white wine
redWine['class'] = 1    # red wine

wineData = pd.concat([whiteWine, redWine], ignore_index=True)

wineData = wineData.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
data_analysis = wineData.drop('class', axis=1)
correlation_matrix = data_analysis.corr()

plt.figure(figsize=(12, 8))  # Ajusta el tama√±o si es necesario
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
class_counts = wineData['class'].value_counts()

plt.figure(figsize=(8, 5))
sns.countplot(data=wineData, x='class')
plt.title('Distribution of Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['White Wine (0)', 'Red Whine (1)'])
plt.show()

class_percentage = (class_counts / len(wineData)) * 100
print("\nClass Percentages:")
print(class_percentage)
```

```python
wineData.to_csv('whine-data/wine_dataset.csv', sep=',', index=False, encoding='utf-8')
```
