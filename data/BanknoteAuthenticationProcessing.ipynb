```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
banknoteData = pd.read_csv('BanknoteAuthentication/data_banknote_authentication.csv')
banknoteData.head()
```

```python
unique_classes = banknoteData['class'].unique()
print("Unique class labels:", unique_classes)
```

```python
# Drop the 'class' column for analysis
data_analysis = banknoteData.drop('class', axis=1)

correlation_matrix = data_analysis.corr()
plt.figure(figsize=(12, 8))

# Create a heatmap for the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
class_counts = banknoteData['class'].value_counts()

plt.figure(figsize=(8, 5))
sns.countplot(data=banknoteData, x='class')
plt.title('Distribution of Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Counterfeit (0)', 'Authentic (1)'])
plt.show()

class_percentage = (class_counts / len(banknoteData)) * 100
print("\nClass Percentages:")
print(class_percentage)
```
